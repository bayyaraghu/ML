```python
# Adding the new column in dataset using function inside if statemetns

def size(n):
    if n<4:
        return "Small"
    elif 4<n<6:
        return "Medium"
    else:
        return "Big"
housing['House_size']=housing['Avg. Area Number of Rooms'].apply(size)

```

```python
import pdpipe as pdp
```

```python
pipeline = pdp.ColDrop('Avg. Area House Age')
pipeline+=pdp.OneHotEncode('House_size')
df3=pipeline(housing)
```
